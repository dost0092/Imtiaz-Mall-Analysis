# 2. Data Cleaning:

### Identify and handle missing values using appropriate techniques like mean/median imputation or dropping rows/columns with excessive missingness.

# A. K-Means Clustering:

## Average_Spending_Per_Purchase'VS 'Purchase_Frequency_Per_Month'VS 'Brand_Affinity_Score'

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Extract relevant columns for clustering
clustering_data = df[['Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score']]

# Standardize the data
scaler = StandardScaler()
clustering_data_standardized = scaler.fit_transform(clustering_data)

# Function to choose the optimal number of clusters using the elbow method
def choose_optimal_clusters_elbow(data, max_clusters=10):
    sse = []

    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
        kmeans.fit(data)
        sse.append(kmeans.inertia_)

    # Identify the optimal number of clusters using the elbow method
    optimal_k_elbow = sse.index(min(sse)) + 1

    # Plot the Elbow method
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, max_clusters + 1), sse, marker='o')
    plt.title('Elbow Method for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Sum of Squared Distances (SSE)')
    plt.show()

    return optimal_k_elbow

# Choose the optimal number of clusters using the elbow method
optimal_k_elbow = choose_optimal_clusters_elbow(clustering_data_standardized)

# Function to choose the optimal number of clusters using silhouette analysis
def choose_optimal_clusters_silhouette(data, max_clusters=10):
    silhouette_scores = []

    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
        kmeans.fit(data)
        silhouette_scores.append(silhouette_score(data, kmeans.labels_))

    # Find the optimal number of clusters with the highest silhouette score
    optimal_k_silhouette = silhouette_scores.index(max(silhouette_scores)) + 2
    return optimal_k_silhouette

# Choose the optimal number of clusters using silhouette analysis
optimal_k_silhouette = choose_optimal_clusters_silhouette(clustering_data_standardized)

# Use the minimum of the optimal values from both methods
optimal_k = min(optimal_k_elbow, optimal_k_silhouette)

# Perform K-Means clustering with the chosen number of clusters
kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
df['KMeans_Cluster_Clean'] = kmeans.fit_predict(clustering_data_standardized)

# Include 'Product_Category_Preferences' in the cluster summary
cluster_summary = df.groupby('KMeans_Cluster_Clean').agg({
    'Average_Spending_Per_Purchase': 'mean',
    'Purchase_Frequency_Per_Month': 'mean',
    'Brand_Affinity_Score': 'mean',
}).reset_index()

# Scatter plot for 'Average_Spending_Per_Purchase' vs 'Purchase_Frequency_Per_Month' using Seaborn
plt.figure(figsize=(10, 5))

# Seaborn scatter plot
sns.scatterplot(data=df, x='Average_Spending_Per_Purchase', y='Purchase_Frequency_Per_Month',
                hue='KMeans_Cluster_Clean', palette='viridis', legend='full')
centers_kmeans = kmeans.cluster_centers_
for i in range(optimal_k):
    plt.gca().add_patch(plt.Circle((centers_kmeans[i, 0], centers_kmeans[i, 1]), 0.4, color='black',
                                   fill=False, linestyle='dashed'))
plt.title(f'Clean KMeans Clustering: Average Spending vs Purchase Frequency (k={optimal_k})')
plt.xlabel('Average Spending per Purchase')
plt.ylabel('Purchase Frequency per Month')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Display cluster summary
print("Cluster Summary:")
```

# B. DBSCAN Clustering:

## Purchase Amount vs Average Spending per Purchas Vs Brand_Affinity_Score 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import seaborn as sns

# Extract relevant columns for clustering
clustering_data = df[['Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score']]

# Standardize the data
clustering_data_standardized = StandardScaler().fit_transform(clustering_data)

# Set parameters for DBSCAN to achieve 6 clusters
eps = 0.5  # Adjust as needed
min_samples = 10  # Adjust as needed

# Perform DBSCAN clustering
df['DBSCAN_Cluster'] = DBSCAN(eps=eps, min_samples=min_samples).fit_predict(clustering_data_standardized)

# Include 'Product_Category_Preferences' in the cluster summary
cluster_summary_dbscan = df.groupby('DBSCAN_Cluster').agg({
    'Average_Spending_Per_Purchase': 'mean',
    'Purchase_Frequency_Per_Month': 'mean',
    'Brand_Affinity_Score': 'mean',
}).reset_index()

# Scatter plot for 'Average_Spending_Per_Purchase' vs 'Purchase_Frequency_Per_Month' using Seaborn
plt.figure(figsize=(10, 5))
sns.scatterplot(data=df, x='Average_Spending_Per_Purchase', y='Purchase_Frequency_Per_Month',
                hue='DBSCAN_Cluster', palette='viridis', legend='full')
plt.title('DBSCAN Clustering: Average Spending vs Purchase Frequency')
plt.xlabel('Average Spending per Purchase')
plt.ylabel('Purchase Frequency per Month')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Display DBSCAN cluster summary
print("DBSCAN Cluster Summary:")
print(cluster_summary_dbscan)
```

# C. K-Means++ Clustering:

## Purchase Amount vs Average Spending per Purchas Vs Brand_Affinity_Score 

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Assuming your DataFrame is named 'df'
# Extract relevant columns for clustering
clustering_data = df[['Average_Spending_Per_Purchase', 'Purchase_Frequency_Per_Month', 'Brand_Affinity_Score']]

# Standardize the data
scaler = StandardScaler()
clustering_data_standardized = scaler.fit_transform(clustering_data)

# Function to choose the optimal number of clusters using the elbow method
def choose_optimal_clusters_elbow(data, max_clusters=10):
    sse = []

    for k in range(1, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
        kmeans.fit(data)
        sse.append(kmeans.inertia_)

    # Identify the optimal number of clusters using the elbow method
    optimal_k_elbow = sse.index(min(sse)) + 1

    # Plot the Elbow method
    plt.figure(figsize=(10, 5))
    plt.plot(range(1, max_clusters + 1), sse, marker='o')
    plt.title('Elbow Method for Optimal K')
    plt.xlabel('Number of Clusters (K)')
    plt.ylabel('Sum of Squared Distances (SSE)')
    plt.show()

    return optimal_k_elbow

# Choose the optimal number of clusters using the elbow method
optimal_k_elbow = choose_optimal_clusters_elbow(clustering_data_standardized)

# Function to choose the optimal number of clusters using silhouette analysis
def choose_optimal_clusters_silhouette(data, max_clusters=10):
    silhouette_scores = []

    for k in range(2, max_clusters + 1):
        kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
        kmeans.fit(data)
        silhouette_scores.append(silhouette_score(data, kmeans.labels_))

    # Find the optimal number of clusters with the highest silhouette score
    optimal_k_silhouette = silhouette_scores.index(max(silhouette_scores)) + 2
    return optimal_k_silhouette

# Choose the optimal number of clusters using silhouette analysis
optimal_k_silhouette = choose_optimal_clusters_silhouette(clustering_data_standardized)

# Use the minimum of the optimal values from both methods
optimal_k = min(optimal_k_elbow, optimal_k_silhouette)

# Perform K-Means clustering with the chosen number of clusters and KMeans++ initialization
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', random_state=42, n_init=10)  # Set n_init explicitly to suppress the warning
df['KMeans_Cluster_Clean'] = kmeans.fit_predict(clustering_data_standardized)

# Include 'Product_Category_Preferences' in the cluster summary
cluster_summary = df.groupby('KMeans_Cluster_Clean').agg({
    'Average_Spending_Per_Purchase': 'mean',
    'Purchase_Frequency_Per_Month': 'mean',
    'Brand_Affinity_Score': 'mean',
}).reset_index()

# Scatter plot for 'Average_Spending_Per_Purchase' vs 'Purchase_Frequency_Per_Month' using Seaborn
plt.figure(figsize=(8, 4))

# Seaborn scatter plot
sns.scatterplot(data=df, x='Average_Spending_Per_Purchase', y='Purchase_Frequency_Per_Month',
                hue='KMeans_Cluster_Clean', palette='viridis', legend='full')
centers_kmeans = kmeans.cluster_centers_
for i in range(optimal_k):
    plt.gca().add_patch(plt.Circle((centers_kmeans[i, 0], centers_kmeans[i, 1]), 0.4, color='black',
                                   fill=False, linestyle='dashed'))
plt.title(f'Clean KMeans Clustering: Average Spending vs Purchase Frequency (k={optimal_k})')
plt.xlabel('Average Spending per Purchase')
plt.ylabel('Purchase Frequency per Month')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()

# Display cluster summary
print("Cluster Summary:")
print(cluster_summary)
```
